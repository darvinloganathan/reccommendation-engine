```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('ratings.csv')
```

```python
data.head()
```

```python
df=pd.pivot_table(data,values='rating',index=['userId'], columns=['movieId'])
df=df.fillna(0)
df
```

```python
from sklearn.metrics.pairwise import pairwise_distances
user_similarity = pairwise_distances(df, metric='cosine')
movie_similarity = pairwise_distances(df.T, metric='cosine')
```

```python
mean_user_rating = df.mean(axis=1)[:, np.newaxis] 
ratings_diff = (df- mean_user_rating) 
user_pred = mean_user_rating + user_similarity.dot(ratings_diff) / np.array([np.abs(user_similarity).sum(axis=1)]).T
user_pred
```

```python
movie_pred = df.dot(movie_similarity) / np.array([np.abs(movie_similarity).sum(axis=1)])
movie_pred
```

```python
similar_to_movie1 = df.corrwith(movie1_user_ratings)
```

```python
corr = pd.DataFrame(similar_to_movie1,columns=['Correlation'])
corr.dropna(inplace=True)
corr=corr.sort_values(by=['Correlation'],ascending=False)
corr.head()
```

```python
def rec_movie(x):
    rate_x=df[x]
    similar_movie=df.corrwith(rate_x)
    corr=pd.DataFrame(similar_movie,columns=['correlation'])
    corr=corr.dropna()
    corr=corr.sort_values(by=['correlation'], ascending=False)
    print(corr.head())
```

```python
rec_movie(10)
```
